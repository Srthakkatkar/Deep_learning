**Simple Deep Neural Network (10 layers)**

```python
import numpy as np

# Input
x = 2.0   # changed input value

# Activation function
def relu(z):
    return max(0, z)

# Initialize weights and biases (changed values)
weights = [0.3, 0.6, 0.4, 0.8, 0.5, 0.9, 0.2, 0.7, 0.4]
biases  = [0.2, 0.1, 0.3, 0.2, 0.1, 0.4, 0.2, 0.3, 0.1]

# Forward propagation through first 9 layers
a = x
for i in range(9):
    z = a * weights[i] + biases[i]
    a = relu(z)
    print(f"Layer {i+1} output:", a)

# Output layer (10th layer)
output = a * 0.6 + 0.3
print("Final Output:", output)
```

**Vanashing GD**

```python
# Input
x = 1.0

def relu(z):
    return max(0, z)

# Very small weights
weights = [0.1] * 9
biases  = [0.01] * 9

a = x
for i in range(9):
    z = a * weights[i] + biases[i]
    a = relu(z)
    print(f"Layer {i+1} output:", a)

# Output layer
output = a * 0.1 + 0.01
print("Final Output (Vanishing):", output)
```

**Exploding Gradient**

```python
# Input
x = 1.0

def relu(z):
    return max(0, z)

# Very large weights
weights = [2.0] * 9
biases  = [1.0] * 9

a = x
for i in range(9):
    z = a * weights[i] + biases[i]
    a = relu(z)
    print(f"Layer {i+1} output:", a)

# Output layer
output = a * 2.0 + 1.0
print("Final Output (Exploding):", output)
```
